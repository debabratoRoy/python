#### Integrals & Derivatives

```python
x = np.linspace(1,10,100)
y = 1/x**2 * np.sin(x)

# derivative
dydx = np.gradient(y,x)

# integral
y_integral = np.cumsum(y) * (x[1]-x[0])
```

```python
plt.plot(x,y,label='$y(x)$',color='#091833')
plt.plot(x,dydx,label=r'$\frac{dy}{dx}$',color='#e34234')
plt.plot(x,y_integral,label='$\int y(x)$',color='#b12dd0')

plt.xlabel('$x$')

plt.legend()
plt.show()
```

#### Question 4:

The power emitted by a blackbody is $P = A\sigma \epsilon T^4$. After measuring the temperature of a star you find that $T(t) = \frac{T_0}{1+e^{-kt}}$. Plot the total energy emitted by the star as a function of time using the fact that $E(t) =\displaystyle \int_{t'=0}^{t'=t} P(t') dt'$

#### Systems of Equations

$
3x + 2y + z = 4 \\
5x - 5y + 4z = 3 \\
6x+z = 0
$

We'd turn the equations in two matrices.

$$
A = 
\begin{bmatrix}
3 & 2 & 1 \\
5 & -5 & 4 \\
6 & 0 & 1 \\
\end{bmatrix}

c = 
\begin{bmatrix}
4 \\ 3 \\ 0
\end{bmatrix}
$$

```python
A = np.array([[3,2,1],[5,-5,4],[6,0,1]])
c = np.array([4,3,0])
# It could be written in the following way too
# c = np.array([[4],[3],[0]])

A,c
```

```python
np.linalg.solve(A,c)
```

#### Finding Eigenvalues

Consider the following matrix,  
$$
A = 
\begin{bmatrix}
4 & 2 & 2 \\
2 & 4 & 2 \\
2 & 2 & 4 \\
\end{bmatrix}
$$

Find its eigenvalues

```python
A = np.array([[4,2,2],[2,4,2],[2,2,4]])

# calculating and then separating eigen values and eigen vectors
# the eigen vectors are in the colums
l, v = np.linalg.eig(A)
l
v # first column is the eigen vector for the first eigen value and so on 
```

```python
v1, v2, v3 = v[:,0], v[:,1], v[:,2]
l1, l2, l3 = l

A @ v1
l1*v1

A @ v2
l2 * v2
```

#### Question 4

Find the solutions to $\left( \frac{d^2}{dx^2} + (10x)^2 \right){f} = \lambda{f}$ with boundary conditions $f(0) = f(1) = 0$

- Use the Eigenvalue Method and the fact that $\frac{d^2f}{dx^2} \approx \frac{f_{i+1} + f_{i-1} - 2f_{i} }{\Delta{x}^2}$ where $f_{i} = f(i\Delta{x})$ and $i$ in an integer.

```python
N = 1000
x = np.linspace(0,1,N+1)
dx = x[1] - x[0]
```

```python
main_diag = -2*np.ones(N-1)
off_diag = np.ones(N-2)

derivative_matrix = ( np.diag(main_diag) + np.diag(off_diag,1) + np.diag(off_diag,-1) ) / dx**2

x2_matrix = np.diag((10*x[1:-1])**2)

LHS_matrix = derivative_matrix + x2_matrix
```

```python
w, v = np.linalg.eigh(LHS_matrix)
```

```python
plt.plot(v[:,-1],label='-1')
plt.plot(v[:,-2],label='-2')
plt.plot(v[:,-3],label='-3')
plt.plot(v[:,-4],label='-4')
plt.legend()
```
